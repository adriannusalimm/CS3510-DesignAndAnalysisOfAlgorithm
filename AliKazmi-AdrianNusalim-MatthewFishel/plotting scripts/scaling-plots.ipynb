```python
#first plot: Performance (“fraction of squares revealed”) vs bomb density
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
fig = plt.figure()
ax1 = fig.add_subplot(111)

densityAlgo1 = pd.read_csv('averaged_density_output_algo1.txt', delimiter = " ")
densityAlgo2 = pd.read_csv('averaged_density_output_algo2.txt', delimiter = " ")

densityAlgo1['fraction_of_squares_revealed'] = densityAlgo1.apply(lambda row: row.dig_count/row.numsquares, axis=1)
densityAlgo2['fraction_of_squares_revealed'] = densityAlgo2.apply(lambda row: row.dig_count/row.numsquares, axis=1)

densityAlgo1['bomb_density'] = densityAlgo1.apply(lambda row: (100*row.bomb_count)/row.numsquares, axis=1)
densityAlgo2['bomb_density'] = densityAlgo2.apply(lambda row: (100*row.bomb_count)/row.numsquares, axis=1)

ax1.scatter(x=densityAlgo1["fraction_of_squares_revealed"], y=densityAlgo1["bomb_density"],c='b',label='Algo 1');
ax1.scatter(x=densityAlgo2["fraction_of_squares_revealed"], y=densityAlgo2["bomb_density"],c='r',label='Algo 2');

plt.yticks(np.arange(0, max(densityAlgo2.bomb_density)+2, 2))
plt.legend(loc='upper left');
plt.title(label="Performance (“fraction of squares revealed”) (dependent) vs bomb density (independent)")
plt.ylabel("bomb density (%) ")
plt.xlabel("fraction of squares revealed (averaged over 5 trials on different boards)")
```
